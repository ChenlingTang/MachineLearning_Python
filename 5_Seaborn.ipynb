```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## set style

def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
```

```python
## the defalut style (darkgrid)
sns.set()
sinplot()
```

Like colorbrewer in R

6个默认的颜色循环主题： deep, muted, pastel, bright, dark, colorblind

## Facetgrid
Display sub-groups data

```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt

sns.set(style="ticks")
np.random.seed(sum(map(ord, "axis_grids")))
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
g = sns.FacetGrid(tips, col="time")
g.map(plt.hist, "tip");
```

```python
## divide data by sex,and color by smoker
g = sns.FacetGrid(tips, col="sex", hue="smoker")
g.map(plt.scatter, "total_bill", "tip", alpha=.7)
g.add_legend();
```

```python
## divide data to four sections
g = sns.FacetGrid(tips, row="smoker", col="time", margin_titles=True)
g.map(sns.regplot, "size", "total_bill",  fit_reg=False, x_jitter=.1);
```

```python
##chane figure size
g = sns.FacetGrid(tips, col="day", size=4, aspect=.5)
g.map(sns.barplot, "sex", "total_bill");
```

```python
## assign the order(row_order)
from pandas import Categorical
ordered_days = tips.day.value_counts().index
print (ordered_days)
ordered_days = Categorical(['Thur', 'Fri', 'Sat', 'Sun'])
g = sns.FacetGrid(tips, row="day", row_order=ordered_days,
                  size=1.7, aspect=4,)
g.map(sns.boxplot, "total_bill");
```

```python
## more detail
pal = dict(Lunch="seagreen", Dinner="gray")
g = sns.FacetGrid(tips, hue="time", palette=pal, size=5)
g.map(plt.scatter, "total_bill", "tip", s=50, alpha=.7, linewidth=.5, edgecolor="white")
g.add_legend();
```

```python
g = sns.FacetGrid(tips, hue="sex", palette="Set1", size=5, hue_kws={"marker": ["^", "v"]})
g.map(plt.scatter, "total_bill", "tip", s=100, linewidth=.5, edgecolor="white")
g.add_legend();
```

```python
## add xticks and yticks
## set figure gap size
with sns.axes_style("white"):
    g = sns.FacetGrid(tips, row="sex", col="smoker", margin_titles=True, size=2.5)
g.map(plt.scatter, "total_bill", "tip", color="#334488", edgecolor="white", lw=.5);
g.set_axis_labels("Total bill (US Dollars)", "Tip");
g.set(xticks=[10, 30, 50], yticks=[2, 6, 10]);
g.fig.subplots_adjust(wspace=.02, hspace=.02);
#g.fig.subplots_adjust(left  = 0.125,right = 0.5,bottom = 0.1,top = 0.9, wspace=.02, hspace=.02)
```

```python
## assign figure kinds to pairplot
g = sns.PairGrid(iris, hue="species")
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
g.add_legend()
```

```python
## select target colmun for pair
g = sns.PairGrid(iris, vars=["sepal_length", "sepal_width"], hue="species")
g.map(plt.scatter);
```

## Heatmap

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np; 
np.random.seed(0)
import seaborn as sns;
sns.set()
```

```python
uniform_data = np.random.rand(3, 3)
sns.heatmap(uniform_data)
```

```python
## set the minimum and max value
sns.heatmap(uniform_data, vmin=0.2, vmax=0.5)
```

```python
## set the center of all value
ax = sns.heatmap(uniform_data, center=0.6)
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
## convert data format
flights = flights.pivot("month", "year", "passengers")
sns.heatmap(flights)
```

```python
## add annotation to each blocks
ax = sns.heatmap(flights, annot=True,fmt="d")
```

```python
## adjust block grid size
ax = sns.heatmap(flights, linewidths=0.1)
```

```python
## chane color
ax = sns.heatmap(flights, cmap="YlGnBu")
```

```python
## hidden color bar
ax = sns.heatmap(flights, cbar=False)
```
